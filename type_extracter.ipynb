```python
import pandas as pd
import numpy as np
import pdb
import re
from os import listdir
from collections import Counter
import pickle
```

```python
data = pd.read_csv("VTRepo_top3_label.csv")
dir_path = "top3_label/trace_by_family_5up"
families = [ll for ll in listdir(dir_path) if ll != ".DS_Store"]
```

```python
label_dict = {}
for i, row in data.iterrows():
    label_dict[row[0]] = [re.sub(r'[^\w\s]',' ',name).strip() for name in list(row[1:].dropna())]
```

```python
fterms = {}
for family in families:
    vids = np.unique([e.split("_")[0] for e in listdir(dir_path + "/" + family)])
    terms = []
    for vid in vids:
        for e in label_dict[vid]:
            terms += [ee.strip() for ee in e.split(" ")]
    fterms[family] = Counter(terms).most_common(10)
```

```python
pickle.dump(fterms, open("fami-type.pkl", "wb"))
```
