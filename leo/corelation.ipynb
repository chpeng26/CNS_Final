```python
def correlation_matrix(df):
    from matplotlib import pyplot as plt
    from matplotlib import cm as cm

    fig = plt.figure()
    ax1 = fig.add_subplot(111)
    cmap = cm.get_cmap('jet', 30)
    cax = ax1.imshow(df.apply(lambda x: x.factorize()[0]).corr(), interpolation="nearest", cmap=cmap)
#     ax1.grid(True)
    plt.title('AV mal_name Correlation')
    labelX = [' ','Symantec','Microsoft','TrendMicro','Kaspersky']
    labels=[' ','Symantec',' ','Microsoft',' ','TrendMicro',' ','Kaspersky']
    ax1.set_xticklabels(labelX,fontsize=9)
    ax1.set_yticklabels(labels,fontsize=11)
    # Add colorbar, make sure to specify tick locations to match desired ticklabels
    fig.colorbar(cax ) #ticks=[0,.75,.8,.85,.90,.95,1]
    plt.show()

correlation_matrix(df_clean)
```
